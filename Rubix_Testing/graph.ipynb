```python
import matplotlib.pyplot as plt
import json

port_logfile_map = {
    20000: "Log/LogPort1.log",
    20001: "Log/LogPort2.log",
    20002: "Log/LogPort3.log",
    20003: "Log/LogPort4.log",
    20004: "Log/LogPort5.log",
    20005: "Log/LogPort6.log",
    20006: "Log/LogPort7.log",
    20007: "Log/LogPort8.log",
    20008: "Log/LogPort9.log",
    20009: "Log/LogPort10.log",
}

for port, logfile in port_logfile_map.items():

    with open(logfile, 'r') as f:
        lines = f.readlines()

    data = [json.loads(line.replace("'", "\"")) for line in lines]

    cpu_usage = [entry['CPU usage'] for entry in data if 'CPU usage' in entry]
    ram_usage = [entry['RAM usage'] for entry in data if 'RAM usage' in entry]

    fig, axs = plt.subplots(2, figsize=(10, 6))

    axs[0].plot(cpu_usage, label=f'CPU usage for port {port}', color='red', linestyle='--')
    axs[0].set_title(f'CPU and RAM usage for port {port}', fontsize=14)
    axs[0].set_ylabel('CPU usage (%)', fontsize=12)
    axs[0].grid(True)
    axs[0].legend()

    axs[1].plot(ram_usage, label=f'RAM usage for port {port}', color='orange')
    axs[1].set_ylabel('RAM usage (MB)', fontsize=12)
    axs[1].set_xlabel('Time (s)', fontsize=12)
    axs[1].grid(True)
    axs[1].legend()

    plt.show()
```
